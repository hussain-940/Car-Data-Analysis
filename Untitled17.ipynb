```python
import pandas as pd
data = {
    'car': ['Toyota', 'Ford', 'Honda', 'BMW'],
    'fuel_type': ['Petrol', 'Diesel', 'Petrol', 'Diesel'],
    'price': [20000, 18000, 22000, 35000]
}
df = pd.DataFrame(data)
average_price_per_fuel = df.groupby('fuel_type')['price'].mean()
max_fuel_type = average_price_per_fuel.idxmax()
max_avg_price = average_price_per_fuel.max()
print(f"The most expensive fuel type is {max_fuel_type} with an average price of {max_avg_price:.2f}")
```

```python
import pandas as pd

data = {
    'car': ['Toyota', 'Ford', 'Honda', 'BMW'],
    'fuel_type': ['Petrol', 'Diesel', 'Petrol', 'Diesel'],
    'price': [20000, 18000, 22000, 35000]
}
df = pd.DataFrame(data)
average_price_per_fuel = df.groupby('fuel_type')['price'].mean()
max_fuel_type = average_price_per_fuel.idxmax()
max_avg_price = average_price_per_fuel.max()
print(f"The most expensive fuel type is {max_fuel_type} with an average price of {max_avg_price:.2f}")
numerical_df = df.select_dtypes(include=['number'])
correlation = numerical_df.corr(method='pearson')
print(correlation)
```
